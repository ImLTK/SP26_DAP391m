```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Dữ liệu giả định
can_nang = np.array([50, 52, 63, 69, 65])
chieu_cao = np.array([152, 153, 165, 172, 167])

data = pd.DataFrame({
    'can_nang': can_nang,
    'chieu_cao': chieu_cao
})

data
```

```python
# shape du lieu dau vao
X = can_nang.reshape(-1, 1)
y = chieu_cao

X
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")
print(f"R-squared: {model.score(X, y)}")
```

```python
# plotting
plt.figure(figsize=(10,6))
plt.scatter(can_nang, chieu_cao, color='yellow', label='Original Data')
plt.plot(can_nang, model.predict(X), color='blue', linewidth=2, label='Linear Regression Line')
plt.xlabel('Cân nặng (KG)')
plt.ylabel('Chiều cao (cm)')
plt.title('Hồi quy tuyến tính cho chiều cao')
plt.legend()
plt.show()
```

```python
# Dự đoán
new_can_nang = np.array([86, 88, 89]).reshape(-1,1)
predicted_height = model.predict(new_can_nang)

print("\nDự đoán chiều cao cho người 86 kg, 88 kg, 89 kg:")
for i, v in zip(new_can_nang.flatten(), predicted_height):
   print(f"Cân nặng {i} kg: {v:.2f} cm")
```
