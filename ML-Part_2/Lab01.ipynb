```python
from chefboost import Chefboost as chef
import pandas as pd
```

```python
df = pd.read_csv("/mnt/golf.txt")
df.head()
```

```python
print("ID3 algorithm\n")

#set algorithm
config = {'algorithm': 'ID3'}

#train model on training dataset
print("Train model on traing dataset:")
model = chef.fit(df[:8].copy(), config)

#define x vector for given day
xrow = ['Sunny','Hot','High','Weak']

#make prediction for given day
prediction = chef.predict(model, xrow)
print("Prediction for given day. Should we play? {0}".format(prediction))

c=0
print("\nCompare predictions and expected results on TESTING dataset:")
#prediction on testing data set
for index, instance in df[8:].iterrows():
    prediction = chef.predict(model, instance)
    actual=instance['Decision']
    print(actual,"-",prediction)
    if(actual==prediction):
        c=c+1
print("Accuracy on TESTING data set:{0}%".format(round(c/df[8:].shape[0],4) ))
```

**CART**

```python
print("CART alogorithm")

config = {'algorithm': 'CART'}
model = chef.fit(df[:8],config)
xrow = ['Sunny', 'Hot', 'High', 'Weak']
c=0
prediction  =  chef.predict(model, xrow)
print("Testing on test dataset")
for index, instance in df[8:].iterrows():
  prediction = chef.predict(model, instance)
  actual = instance['Decision']
  print(actual,'-',prediction)
  if(actual==prediction):
        c=c+1
print(f'Accuracy on testing set: {round(c/df[8:].shape[0],4)}')
```

**C5.0**

```python
from sklearn.tree import DecisionTreeClassifier
print('C5.0')

df_encoded = pd.get_dummies(df.drop('Decision', axis = 1))
y = df['Decision']
c=0
clf = DecisionTreeClassifier(criterion='entropy', max_depth = 3)
clf.fit(df_encoded, y)

X_test_encoded = df_encoded[8:]
y_test = y[8:]

print(f'Testing on test dataset')
for index, instance_features in X_test_encoded.iterrows():
  pred = clf.predict(instance_features.values.reshape(1, -1))[0]
  actual = y_test.loc[index]
  if actual == pred:
        c = c + 1
  print(f"{actual} - {pred}")
print(f'Accuracy on test ddata set: {round(c/X_test_encoded.shape[0],4)}')
```
