import data from the 'titanic.csv' file and check its structure.

passenger_id – Unique passenger id

pclass – Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd)

survived – Survival (0 = No, 1 = Yes)

```python
titanic = pd.read_csv("/mnt/titanic.csv")
```

```python
titanic.info()
```

```python
# data understanding
titanic.sample(10)
```

```python
# passenger_id – Unique passenger id
print(f"Unique ids: {titanic["passenger_id"].unique()}")
print(f"Unique ids count: {titanic["passenger_id"].value_counts()}")
```

```python
# pclass – Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd)
print(f"Ticket classes: {titanic.pclass.value_counts().sort_values(ascending = False)}")
```

```python
# survived – Survival (0 = No, 1 = Yes)
print(f"Unique data points: {titanic.survived.unique()}")
print(f"Survived: {titanic.survived.value_counts()}")
```

```python
# name – Name and SUrname
print(f"Names and surnames: {titanic.name.value_counts()}")
```

```python
# sex – Sex (0 = Male, 1 = Female)
sex = titanic
print(f"Sex: {sex.sex.map({0: "Male", 1: "Female"}).value_counts()}")
```

```python
# age – Age in years
print(f"Ages in years: ")
print(titanic.age.value_counts())
print("Unique Ages: ")
print(titanic.age.fillna(0).astype(int).unique())
```

```python
# sibsp – # of siblings / spouses aboard the Titanic
print(f"Number of siblings / spouses that aboard the Titanic: {titanic.sibsp.sum()}")
```

```python
# parch – # of parents / children aboard the Titanic
print(f"Number of parents / children aboard the Titanic: {titanic.parch.sum()}")
```

```python
# ticket – Ticket number
print(f"Unique ticket numbers: ")
print(titanic.ticket.unique())
print(f"There are {titanic.ticket.nunique()} tickets")
```

```python
#  fare – Passenger fare
print("Passengers fare: ")
print(titanic.fare.value_counts())
print("Total fare: ")
print(titanic.fare.sum())
```

```python
# cabin – Cabin number
print(f"Unique cabin numbers: {titanic.cabin.unique()}")
```

```python
# embarked – Port of Embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)
print(titanic.embarked.map({"C": "Cherbourg", "Q": "Queenstown", "S": "Southampton"}).value_counts())
```

```python
# boat – Lifeboat (if survived)
print(titanic.boat.value_counts())
```

```python
# body – Body number (if did not survive and body was recovered)
print(f"Total bodies: {titanic.body.sum().astype(int)}")
```

```python
print("All home/destination: ")
print(titanic["home.dest"])
```

In the titanic dataset convert columns with qualitative variables to factors (sex, embarked, survived) or ordered factor (pclass) - defining the appropriate levels and, if needed, order of values. Please code the levels as:

embarked respectively: Cherbourg, Queenstown, Southampton

sex - Female, Male

survived - Yes, No

pclass - 1st <2nd <3rd

```python
new_titanic = titanic
```

```python
new_titanic.sample(10)
```

```python
# embarked respectively: Cherbourg, Queenstown, Southampton
new_titanic["embarked"] = new_titanic["embarked"].map({"C": "Cherbourg", "Q": "Queenstown", "S": "Southampton"})
new_titanic["embarked"].unique()
```

```python
# sex - Female, Male

new_titanic["sex"] = new_titanic["sex"].map({0: "Male", 1: "Female"})
new_titanic["sex"].unique()
```

```python
# survived

new_titanic["survived"] = new_titanic["survived"].map({0: "No", 1: "Yes"})
new_titanic.survived.unique()
```

```python
# pclass - 1st <2nd <3rd
new_titanic = new_titanic.sort_values(by = "pclass", ascending=False)
new_titanic.head(10)
```
